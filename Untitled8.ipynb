```python
df=pd.read_csv("imdb.csv")
```

```python
print(df.head())
print(df.info())
print(df.describe())
df.columns = df.columns.str.lower()
print(df['genre'].value_counts())

```

```python
df.isnull().sum()
df.dropna()
df['rating_from_10'] = df['rating_from_10'].fillna(df['rating_from_10'].mean())
```

```python
df.drop_duplicates()
print(df['genre'].unique())
```

```python
sns.boxplot(x=df['rating_from_10'])
plt.show()
```

```python
df.to_csv("cleaned_movies_ratings.csv", index=False)
```

```python
# data exploration
```

```python
sns.histplot(df['rating_from_10'], bins=10, kde=True)
plt.title("Distribution of Movie Ratings")
plt.xlabel("rating_from_10")
plt.ylabel("Count")
plt.show()
```

```python
all_genres = df['genre'].str.split(',').explode().str.strip()
genre_counts = all_genres.value_counts()
```

```python

sns.countplot(y=all_genres, order=genre_counts.index, palette="viridis")
plt.title("Movie Count by Genre")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```

```python
sns.lineplot(x=df['year'], y=df['rating_from_10'], marker="o")
plt.title("Movie Ratings Over Time")
plt.xlabel(" Year")
plt.ylabel("rating_from_10")
plt.show()
```

```python
avg_rating_per_genre = df.groupby('genre')['rating_from_10'].mean().sort_values(ascending=False)
print(avg_rating_per_genre)
```

```python
top_movies = df.nlargest(5, 'rating_from_10')
print(top_movies)
```
